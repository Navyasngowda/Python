```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,5])
plt.bar(x,y,width=0.5)
plt.xlabel('x-axis',fontsize=14)
plt.ylabel("Y-axis",fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
dataSet1=np.random.normal(10,10,20)
dataSet2=np.random.normal(10,10,20)
dataSet3=np.random.normal(10,10,20)
dataSet4=np.random.normal(10,10,20)
dataset=[dataSet1,dataSet2,dataSet3,dataSet4]
figure=plt.figure(figsize=(10,7))
ax=figure.add_axes([1,1,1,1])
bp=ax.boxplot(dataset)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig,ax=plt.subplots(1,1)
a=np.array([22,24,25,26,27,28,28,30])
ax.hist(a,bins=[0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel("no.of students")
plt.show()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
data=[{3,9,3,4},{2,1,9,1},{7,4,3,7},{7,2,3,9}]
Index=['I1','I2','I3','I4']
Cols=['C1','C2','C3','C4']
df=DataFrame(data,index=Index,columns=Cols) 
plt.pcolor(df)
plt.show()
```

```python
'''pi-chart'''
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,55,45])
mylabels=["apples","banansa","oranges","cherries"]
plt.pie(y,labels=mylabels,startangle=10,)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={'year':[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],
     'unemployment_rate':[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
     }
df=pd.DataFrame(data)
plt.plot(df['year'],df['unemployment_rate'],color='blue',marker='p')
plt.title('unemployment rate vs year',fontsize=14,color='green')
plt.xlabel('year',fontsize=14,color='red')
plt.ylabel("unemployment_rate",fontsize=14,color='red')
plt.grid(True)
plt.show()

```
