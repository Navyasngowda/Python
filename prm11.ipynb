```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("iris")
sns.lineplot(x="sepal_length",y="sepal_width",data=data) #draw lineplot
plt.xlim(5) #setting x limit
plt.figure(figsize=(2,4)) #change figsize
sns.set_context("paper")  #setting scale of plot
plt.title("Title using matplotlib function")
sns.set_style("dark")  #changing theme to dark
sns.despine() #removing spines
plt.show()
```

```python
#line plot
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("iris")
sns.lineplot(x="sepal_length",y="species",data=data)
plt.show() #draw lineplot
```

```python
#scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("iris")
sns.scatterplot(x="sepal_length",y="sepal_width",data=data) #draw lineplot
plt.show()
```
